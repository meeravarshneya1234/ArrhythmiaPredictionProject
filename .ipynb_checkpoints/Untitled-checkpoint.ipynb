```python
# Input Data 
filename = 'Manuscript_Data/Population_Feature_Outputs.csv' # file with data 
label = 'IKrBlock_Label'
features = ['Vrest', 'Upstroke', 'Vpeak', 'APD20', 'APD40', 'APD50', 'APD90',
       'TriAP','DCai', 'Capeak', 'CaD50', 'CaD90', 'TriCa', 'dCa']
seed = 134556
```

```python
ML_Task1 = ML_Pipeline(filename,label,features)
df = ML_Task1.convert_df()
```

```python
split_data = train_test_split(df[features], df[label], random_state = seed, test_size = 0.10)
```

```python
LR = ML_Task1.run_LR(split_data)
```

```python
SVM = ML_Task1.run_SVM(split_data)
```

```python
KNN = ML_Task1.run_KNN(split_data)
```

```python
RF = ML_Task1.run_RF(split_data)
```

```python
NB = ML_Task1.run_NB(split_data)
```

```python
GB = ML_Task1.run_GB(split_data)
```

```python
XGB = ML_Task1.run_XGB(split_data)
```

```python
ANN = ML_Task1.run_ANN(split_data)
```

```python
classifiers = [LR[1], SVM[1], NB, XGB[1], GB[1], RF[1], KNN[1], ANN[1]]
keys = ['LR','SVM','NB','XGB','GB','RF','KNN','ANN']
combined_ML_data = ML_Task1.create_table(classifiers,keys)

result_table = combined_ML_data[0]
metrics = combined_ML_data[1]
rocs = combined_ML_data[2]
prob = combined_ML_data[3]
pred = combined_ML_data[4]
conf_matrix = combined_ML_data[5]

result_table
```
